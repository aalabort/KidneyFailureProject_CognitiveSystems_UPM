```r
cleanDrugNames<-function(x)
{
    name = ""
    fields = strsplit(str_trim(tolower(gsub("\\s+"," ",x)))," ")[[1]]
    length(fields)
}
```

```r
cleanDrugNames<-function(x)
{
    name = ""
    x2 = gsub("\\.+","",gsub("\\s+"," ",x)) #Deleting multiple spaces and dots.
    #x2 = gsub("\\s+"," ",x)
    fields = strsplit(str_trim(tolower(x2))," ")[[1]]
    n = length(fields)
    for (i in 1:n){
        if (!is.na(suppressWarnings(as.numeric(substr(fields[i],1,1))))){
            return (str_trim(name))
        }
        name = paste(name,fields[i],sep = "")

    }
    return (str_trim(name))


}
```
