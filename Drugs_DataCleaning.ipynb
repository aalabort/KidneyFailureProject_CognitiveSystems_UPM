```r
cleanDrugNames<-function(x)
{
    name = ""
    x2 = gsub("\\.+","",gsub("\\s+"," ",x)) #Deleting multiple spaces and dots.
    #x2 = gsub("\\s+"," ",x)
    fields = strsplit(str_trim(tolower(x2))," ")[[1]]# Lowering all data, trim whitespace from start and end of string, and split by spaces.
    n = length(fields)# number of words of the element
    for (i in 1:n){
        if (i>1){#Case for 0meprazol
            if (!is.na(suppressWarnings(as.numeric(substr(fields[i],1,1))))) # Keep just the words that do not start by numbers. We will remove the dosis information since it is not very clear and it will dificult to group same drugs under the same name.
            {
                return (str_trim(name))
            }
        }
        name = paste(name,fields[i],sep = "")
    }
    return (str_trim(name))
}
```

```r
drugs_col = colnames(drugs)[colnames(drugs)!='patient_id']
drugs_col
```

```r
drugs_clean = drugs
drugs_clean[drugs_col] <- data.frame(sapply(drugs_clean[drugs_col],function(x) {sapply(x,cleanDrugNames)}),stringsAsFactors=FALSE)
#drugs_cleanX <- data.frame(sapply(drugs[drugs_col],function(x) {sapply(x,cleanDrugNames)}))
#drugs_cleanX$patient_id <- (drugs$patient_id)
#WHY IS DIFERENT -> drugs_clean[drugs_col] <- (apply(drugs_clean[drugs_col],1,function(x) {sapply(x,cleanDrugNames)}))

class(drugs_clean)
```
